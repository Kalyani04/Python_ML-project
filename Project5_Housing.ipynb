```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataframe = pd.read_csv('Housing.csv')
dataframe = dataframe[['price', 'lotsize', 'bedrooms', 'bathrms', 'stories',  'garagepl', 'driveway', 'recroom', 'fullbase', 'gashw', 'airco', 'prefarea' ]]
dataframe = dataframe.replace(to_replace ="yes", value ="1")
dataframe = dataframe.replace(to_replace ="no", value ="0")
print(dataframe)
```

```python
#data = dataframe[[ 'driveway', 'recroom', 'fullbase', 'gashw', 'airco', 'prefarea' ]]
#This is the another method to covert/replace our data (yes or no) into 0 and 1
coverted_driveway = {'driveway': dataframe['driveway']}
df = pd.DataFrame(coverted_driveway,columns=['driveway'])
df.loc[df['driveway'] == 'yes'] = '1' 
df.loc[df['driveway'] == 'no'] = '0' 

print(df) 
converted = np.array(df)
#print(converted)


coverted_recroom = {'recroom': dataframe['recroom']}
df1 = pd.DataFrame(coverted_recroom,columns=['recroom'])
df1.loc[df1['recroom'] == 'yes'] = '1' 
df1.loc[df1['recroom'] == 'no'] = '0' 
print(df1)

coverted_fullbase = {'fullbase': dataframe['fullbase']}
df2 = pd.DataFrame(coverted_fullbase,columns=['fullbase'])
df2.loc[df2['fullbase'] == 'yes'] = '1' 
df2.loc[df2['fullbase'] == 'no'] = '0' 
print(df2)

coverted_gashw = {'gashw': dataframe['gashw']}
df3 = pd.DataFrame(coverted_gashw,columns=['gashw'])
df3.loc[df3['gashw'] == 'yes'] = '1' 
df3.loc[df3['gashw'] == 'no'] = '0' 
print(df3)

coverted_airco = {'airco': dataframe['airco']}
df4 = pd.DataFrame(coverted_airco,columns=['airco'])
df4.loc[df4['airco'] == 'yes'] = '1' 
df4.loc[df4['airco'] == 'no'] = '0' 
print(df4)

coverted_prefarea = {'prefarea': dataframe['prefarea']}
df5 = pd.DataFrame(coverted_prefarea,columns=['prefarea'])
df5.loc[df5['prefarea'] == 'yes'] = '1' 
df5.loc[df5['prefarea'] == 'no'] = '0' 
print(df5)


```

```python
# Create X label
#recroom
input = np.array(dataframe.drop(['price'], 1))
print("Input data",input)
# Create Y label
output = np.array(dataframe.drop([ 'stories'], 1))
print("Output data",output)
```

```python
'''
Select algorithm based on data analysis
- Split training and testing data
- we will be using Linear regression as regression problem
- Tranining data 70%
- Testing data 30%
'''

X_train, X_test, Y_train, Y_test = train_test_split( input, output, test_size = 0.3 )
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)

# Apply Linear Regression
reg = LinearRegression()
model = reg.fit(X_train, Y_train)
```

```python
slope = reg.coef_[0]
intercept = reg.intercept_
print("Slope : {}, Intercept : {}".format(slope, intercept))
```

```python
house_price_prediction_train = np.zeros(Y_train.shape)
float_array = X_train.astype(float)
#print(float_array)
for i in range(len((X_train))):
    house_price_prediction_train[i] = (slope[0] * float_array[i][0])  + intercept
print("predicted training data : {}".format(house_price_prediction_train))
```

```python
plt.plot(house_price_prediction_train, color = 'green')
plt.xlabel('Input data')
plt.ylabel('Price')
plt.plot(dataframe['price'],color = "red")
plt.show()
```

```python
house_price_prediction_test = np.zeros(Y_test.shape)
float_array_test = X_test.astype(float)

for i in range(len((X_test))):
    house_price_prediction_test[i] = (slope[0] * float_array_test[i][0])  + intercept
print("predicted testing data : {}".format(house_price_prediction_test))
```

```python
plt.plot(house_price_prediction_test, color = 'green')
plt.xlabel('Input data')
plt.ylabel('Price')
plt.plot(dataframe['price'],color = "red")
plt.show()
```

```python
a = reg.score(X_train,Y_train)
b = reg.score(X_test,Y_test)
print("Training data : {}, Testing data : {}".format(a,b))
```
